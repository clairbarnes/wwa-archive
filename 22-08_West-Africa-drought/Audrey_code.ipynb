```r
## -----------------------------
##    CALCULATION : onsets, demise and duration of annual rainy season
##    Input_file : daily precipitation 1-D timeseries from year1 to year2 (named file$var here)
## -----------------------------

year1 = 1981     #depends on the timeseries
year2 = 2021    #depends on the timeseries

years=c(year1:year2)    
bisext = seq(1940,2020,4)

onsets=c();ends=c()
start_year=140        #tests between day90 (end of march) to day150 (end of may)
fixed_start=start_year
```

```r
for (ind_all_year in c(1:length(years))){    #loop over years
    year=years[ind_all_year]

    if(c(year %in% bisext)==FALSE){
        end_year = start_year+c(365-1-fixed_start)
    } else if(c(year %in% bisext)==TRUE) {
        end_year = start_year+c(366-1-fixed_start)
    }
    #print(length(c(start_year:end_year)))     #for verif

    ## --- Daily data per year
    tmp_year = file$var[start_year:end_year]

    ## --- Anomalies of precipitation
    S_year = tmp_year-mean(tmp_year)

    ## --- Cumul of rainfall anomalies
    tmp_Syear=c()
    for(i in 1:length(S_year)){
        tmp_Syear[i] = sum(S_year[1:i])
    }    

    ## --- Smoothing anomalies to avoid misleading inflection (e.g. 1 isolated peak)
    ###
    ###
    WHAT_SMOOTH=7.5        #tests were conducted from 5days to 30days smoothing
    ###
    ###
    smoothed_Syear=c()    
    for(i in WHAT_SMOOTH:length(tmp_Syear)){
       smoothed_Syear[i] = mean(tmp_Syear[c(i-WHAT_SMOOTH):c(i+WHAT_SMOOTH)])
    }

    ## --- Derivative of smoothed curve to determine inflection
    deriv_year=c()
    for(i in 1:length(smoothed_Syear)){
        deriv_year[i] = smoothed_Syear[i+1]-smoothed_Syear[i]
    }

    ## --- First (last) day when dérivée >=0 determines the start (end) of the rainy season
    onsets[ind_all_year] = which(deriv_year>=0)[1]
    ends[ind_all_year] = which(deriv_year>=0)[length(which(deriv_year>=0))]

if(c(year %in% bisext)==FALSE){
start_year = start_year+365
}else if(c(year %in% bisext)==TRUE){
start_year = start_year+366
}

}    #end of loop over years

## --- Timeseries from year1 to year2
onsets = onsets+fixed_start
ends = ends+fixed_start
duration = ends-onsets
## -----------------------------
```
