```python
import sys; sys.path.append('/home/clair/wwa'); from wwa import *
cproj = cartopy.crs.PlateCarree()
xn,xx,yn,yx = [-10,0,36,44]

px = {k : get_latlon(k) for k in ["Madrid"]}
```

```python
pr = wrap_lon(xr.open_mfdataset("*.nc")).sel(lon = slice(xn,xx), lat = slice(yn,yx), time = slice("2023-09-02", "2023-09-05")).precipitation.load()
```

```python
pr.plot(col = "time", col_wrap = 8, cmap = "YlGnBu", subplot_kws = {"projection" : cproj, "extent" : [xn,xx,yn,yx]})

for ax in plt.gcf().axes[:-1]: 
    ax.coastlines()
    ax.scatter(px["Madrid"]["lon"], px["Madrid"]["lat"], marker = "x", color = "darkred")
```
