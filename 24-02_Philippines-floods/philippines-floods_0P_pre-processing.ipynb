```python
amo = decode_times(xr.open_dataset("iamo_ersst.nc", decode_times = False)).AMO

amo_feb = amo.rolling(time = 12).mean().groupby("time.month")[2]

amo_feb = amo_feb.assign_coords(time = amo_feb.time.dt.year).rename(time = "year").dropna("year", "any")
```

```python
amo_feb.to_pandas().to_csv("amo.txt", sep = " ")
```

```python
pdo = decode_times(xr.open_dataset("ipdo_ersst.nc", decode_times = False)).index.dropna("time", "any")
```

```python
pdo_ann = pdo.resample(time = "AS-JUL").mean()
```

```python
pdo_ann = pdo_ann.assign_coords(time = pdo_ann.time.dt.year).rename("pdo", time = "year")
```

```python
pdo_ann.to_dataframe().to_csv("pdo.csv")
```

Alternative PDO (more recently updated) from https://oceanview.pfeg.noaa.gov/erddap/tabledap/cciea_OC_PDO.htmlTable?time,PDO

```python
pdo = pd.read_csv("pdo.txt", sep = "\t", parse_dates = ["time"], index_col = "time")

pdo_ann = pdo.resample("AS-JUL", label = "right").mean()

pdo_ann.index = pdo_ann.index.year
pdo_ann.to_csv("pdo.csv")
```

```python
mjo = pd.read_csv("mjo.txt", sep = " +", engine = "python", index_col = "PENTAD", parse_dates = ["PENTAD"], na_values = "*****")

mjo_seas = mjo[["INDEX_5"]].resample("QS-DEC").mean()

mjo_djf = mjo_seas.loc[mjo_seas.index.month == 12]

mjo_djf.index = mjo_djf.index.year+1
mjo_djf.to_csv("mjo5.csv")
```
